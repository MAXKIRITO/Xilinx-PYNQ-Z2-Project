```python
##### -----------------------------Data Pre-Process-----------------------------

dataStr = str(ser.readline())

replaceStrs = ["\\xa4","\\x03","\\tB","\\t","\\n","\\tB"]

for i in replaceStrs:
    dataStr = dataStr.replace(i,"")

print(dataStr)
    
temp1D = dataStr.replace("\\x"," 0x").split(' ')[1:]

temp1D[-1] = temp1D[-1].replace("'","")
print(temp1D)

for i in range(len(temp1D)):
    temp1D[i] = int(temp1D[i],16)
    
temp1D = temp1D[1:65]
print(temp1D)
temp2D = []
for i in range(8):
    temp2D.append(temp1D[0+i:7+i])
for i in temp2D:
    print(i)
    
print("\n-----Data Pre-Process finish-----\n")

##### -----------------------------Get High Temperature Area-----------------------------
highArea = []

for y in range(len(temp2D)):
    for x in range(len(temp2D[y])):
        if temp2D[y][x] >= 20:
            highArea.append((x,y))
            
print(highArea)

print("\n-----Get High Temperature Area finish-----\n")
```
